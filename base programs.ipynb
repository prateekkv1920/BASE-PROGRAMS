# Write a python program for arithmetic calculations. The program should ask for two integers,  and , from the user. The program should compute and display the following:
The sum of  a and b

The difference when b is subtracted from a

The product of a and b

The quotient when a is divided by b

The remainder when a is divided by b

The result of log10 a

The result of a^b

```python
import math
while True:
    choice=input("enter choice-addition, subtract, product, divide, floor_division, log10a, power")
    if choice in ('addition','subtract','product','divide','floor_division','log10a','power'):
        try:
            a=float(input("enter the a integer"))
            b=float(input("enter the b integer"))
        except:
            print("invalid input")
            continue
        if choice=='addition':
            print("addition ",a+b)
        elif choice=='subtract':
            print('subtract',a-b)
        elif choice=='product':
            print('product',a*b)
        elif choice=='divide':
            print('divide',a/b)
        elif choice=='floor_division':
            print('floor_division',a/b)
        elif choice=='log10a':
            print('log10a',math.log10(a))
        elif choice=='power':
            print('power',a**b)
        next_calculation = input("Let's do next calculation? (yes/no): ")
        if next_calculation == "no":
          break
    else:
        print("Invalid Input")   
```

# A zoo charges the admission fee based on the age of the guest. The admission charges for infants of age 2  years or less are not applicable. The guests between  3 and  12 years of age will be charged INR 500. For senior citizens aged 65 years and above the charges are INR 600. For all other citizens, the charges are INR 750. Write a program that prompts the user to enter the age of each guest in a group with one age entered on each line. Then your program should display the number of guests and total admission cost in two decimal places of the group with an appropriate message.

```python
tc = 0
tg =0 
while True:
    age = input("Enter the age of the guest (or 'done' to finish): ")
    if age.lower() == 'done':
        break
    else:
        age = int(age)
        tg += 1
        if age <= 2:
            cost = 0
        elif 3 <= age <= 12:
            cost = 500
        elif age >= 65:
            cost = 600
        else:
            cost = 750
        tc += cost
print(f"\nTotal Guests: {tg}")
print(f"Total Admission Cost: INR {tc:.2f}")
```
